```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#To see how many values are null in the columns we do this

df.isnull().sum()
```

```python
#To see how many unique values are in the columns we do this

df.nunique()
```

```python
#To order our data and see the top 5 countries by 2022 population we can do this

df.sort_values(by='2022 Population', ascending=False).head()
```

```python
#To see the top 10 we can specify it in head(10) and we can try with World Population Percentage

df.sort_values(by='World Population Percentage', ascending=False).head(10)
```

```python
#To see the correlations of the data we use this, is important to do that only for the numeric values to avoid errors

df.corr(numeric_only = True)
```

```python
#To make a better analysis of the correlation is a good idea to make a graphic to see better those correlations

sns.heatmap(df.corr(numeric_only = True), annot = True)  #here you say you want a heatmap of correlations with annotations

plt.rcParams['figure.figsize'] = (30,7) #here you adjust the size of the plot

plt.show()
```

```python
#Now lets group by continent and get the averages

df.groupby('Continent').mean(numeric_only = True)
```

```python
# And now we can make a top of this by 2022 population

df.groupby('Continent').mean(numeric_only = True).sort_values(by = '2022 Population', ascending = False)
```

```python
#To see an specific information we can make a filter on this way

df[df['Continent'].str.contains('Oceania')] 

#This says I want you to show me the df where the column Continent contains the word Oceania
```

```python
#Lets try to visualize data

df2 = df.groupby('Continent').mean(numeric_only = True).sort_values(by = '2022 Population', ascending = False)

df2.plot()
```

```python
#It does not visualize on the way we wanted lets see if we can transpose the data to see better the information

df3 = df2.transpose()
df3
```

```python
df3.plot()

#This does not shows on the ccorrect way because there is more information of different types, lets correct that
```

```python
#We are creating a new table of data in where we speccify the columns that we want on that

df.columns
```

```python
df4 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only = True).sort_values(by = '2022 Population', ascending = False)

df4
```

```python
#other way to obtain the same result would be

df.groupby('Continent')[df.columns[5:13]].mean(numeric_only = True).sort_values(by = '2022 Population', ascending = False)
```

```python
df5 = df4.transpose()
df5
```
