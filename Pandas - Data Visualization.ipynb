Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\chris\OneDrive\Desktop\Python Basics\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
#to create a line plot we use this function by default

df.plot()
```

```python
#We can specify how we want our plot

df.plot(kind = 'line')
```

```python
#We can divide the graphic by each of the elements

df.plot(kind = 'line', subplots = True)
```

```python
#We can customize the title of the plot and the edges making this

df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
#Now lets see a bar chart

df.plot(kind = 'bar')
```

```python
#also we can make an staked bar chart

df.plot(kind = 'bar', stacked = True)
```

```python
#We can specify what column of data we want on the chart

df['Flavor Rating'].plot(kind = 'bar', stacked = True, title = 'Rating by Flavor', xlabel = 'Date', ylabel = 'Score')
```

```python
#And we can make this bar char horizontal putting plot.barh and deleting kind = 'Bar'

df.plot.barh(title = 'Rating by Flavor', xlabel = 'Date', ylabel = 'Score')
```

```python
#For the scatter plot we need to specify the columns for each of the axis, the s is for the size of the points
# and the c is for the color of the dots

df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 200, c = 'Red')
```

```python
#Lets see the histograms and is important to give a value for the bins, to have a more thinner bin you need to increase the value

df.plot.hist(bins = 50)
```

```python
#There are boxplots too

df.boxplot()
```

```python
#An area plot

df.plot.area()
```

```python
#We can adjust the size of the plots with figsize = (x,y)

df.plot.area(figsize = (10,5))
```

```python
#And for the pie chart you need to specify the column to use usig y, example: y='column1'

df.plot.pie(y = 'Flavor Rating')
```

```python
#you can change the style of the visualizations, first lets gonna consult the styles

print(plt.style.available)
```

```python
#And for use the style we do this, remember this changes the style of all the charts when we start that comand

plt.style.use('grayscale')
```
